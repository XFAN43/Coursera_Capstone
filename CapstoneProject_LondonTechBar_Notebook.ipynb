# Introduction

**Background**\
Everyone loves London. London has a long history and is now the financial and technological capital of Europe. The pace of work and life in London is very fast, especially in Central London, which is made of ‘Western Central London’ and ‘Eastern Central London’ and where most of the top schools, financial institutions, international enterprises in London cluster there. 

The advancement of science and technology has brought convenience to people but also great pressure. **2) Target people cluster:** People there have high income/ customer consumption level, but also with high working press\
**3) Accessibility:** Convenient access for people who work nearby\
**4) Culture of having a drink after work:** Several pubs or bars already existed, so people have habits to drink and relax there

**Who Would Be Interested In**\
Businessmen or investors who want to operate or invest a futuristic high-tech bar in Central London.

### Explore all locations in the Central London

```python
# define a getNearbyVenues function
def getNearbyVenues(locations, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for location, lat, lng in zip(locations, latitudes, longitudes):
        print(location)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            location, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    central_nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    central_nearby_venues.columns = ['Location', 
                  'Location Latitude', 
                  'Location Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(central_nearby_venues)
```

```python
# input the coordinates of all the locations in Central London, create a new dataframe called city_venues.

central_venues = getNearbyVenues (locations = central_data['Location'],
                                latitudes=central_data['Latitude'],
                                   longitudes=central_data['Longitude']
                                  )
```

## Analyze venues in Central London

### Transform venues info into dummy variables for further calculating frequency

```python
# one hot encoding
central_onehot = pd.get_dummies(central_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
central_onehot['Location'] = central_venues['Location'] 

# move neighborhood column to the first column
fixed_columns = [central_onehot.columns[-1]] + list(central_onehot.columns[:-1])
central_onehot = central_onehot[fixed_columns]

central_onehot.head()
```

```python
central_onehot.shape
```

### Group rows by Locations and  by taking the mean of the frequency of occurrence of each category

```python
central_grouped = central_onehot.groupby('Location').mean().reset_index()
central_grouped
```

```python
# print each location along with the top 5 most common venues
num_top_venues = 5

for hood in central_grouped['Location']:
    print("----"+hood+"----")
    temp = central_grouped[central_grouped['Location'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

### Let's put that into a pandas dataframe

```python
# define a function

def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
# input to display the top 10 venues for each location.
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Location']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
central_venues_sorted = pd.DataFrame(columns=columns)
central_venues_sorted['Location'] = central_grouped['Location']

for ind in np.arange(central_grouped.shape[0]):
    central_venues_sorted.iloc[ind, 1:] = return_most_common_venues(central_grouped.iloc[ind, :], num_top_venues)

central_venues_sorted.head()
```

```python
central_venues_sorted.shape
```

## Cluster locations in Central London

### Run *k*-means to cluster the locations into 5 clusters.

```python
# set number of clusters
kclusters = 4

central_grouped_clustering = central_grouped.drop('Location', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(central_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:18] 
```

### Create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.


```python
# add clustering labels
central_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
central_merged = central_data

# merge central_grouped with central_data to add latitude/longitude for each location
central_merged = central_merged.join(central_venues_sorted.set_index('Location'), on='Location')

central_merged.head()
```

```python
central_merged.shape
```

### Visualize the resulting clusters

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=12)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(central_merged['Latitude'], central_merged['Longitude'], central_merged['Location'], central_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
